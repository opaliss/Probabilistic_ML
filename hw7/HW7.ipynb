```python
# training data
A = np.loadtxt('hw7files/transitionMatrix.txt', float)
B = np.loadtxt('hw7files/emissionMatrix.txt', float)
pi_1 = np.loadtxt('hw7files/initialStateDistribution.txt', float)
O = np.loadtxt('hw7files/observations.txt', int)
```

```python
print(np.shape(A))
print(np.shape(B))
print(np.shape(pi_1))
print(np.shape(O))

# number of hidden states
n = 27 
T = len(O)
```

```python
O
```

```python
print(np.shape(np.where(O == 0)))
print(np.shape(np.where(O == 1)))
```

```python
def get_emission_prob(idx, state, B=B):
    if state == 1:
        return B[idx, 1]
    if state == 0:
        return B[idx, 0]
```

```python
L = np.zeros((n, T))
```

```python
for t in range(T):
    for jj in range(n):
        if t == 0:
            L[jj, t] = np.log(pi_1[jj]) + np.log(get_emission_prob(idx=jj, state=O[t]))
        elif t > 0:
            L[jj, t] = np.max(L[:, t-1] + np.log(A[:, jj])) + np.log(get_emission_prob(idx=jj, state=O[t]))
```

```python
S = np.zeros(T)
```

```python
for t in range(T):
    if t == 0:
        S[- t - 1] = np.argmax(L[:, - 1])
    elif t: 
        S[- t - 1] = np.argmax(L[:, -t -1] + np.log(A[:, int(S[- t])]))
```

```python
char = ['a','b','c','d','e','f','g','h','i','j','k','l',
               'm','n','o','p','q','r','s','t','u','v','w','x','y','z', ' ']
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.plot(S)
_ = ax.set_yticks(np.arange(0, 27))
_ = ax.set_yticklabels(char)
_ = ax.set_xlabel("time index")
```

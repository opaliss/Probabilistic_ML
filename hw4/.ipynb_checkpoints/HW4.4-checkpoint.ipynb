```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
data_train = np.loadtxt('nasdaq00.txt', float)
data_test = np.loadtxt('nasdaq01.txt', float)

```

```python
plt.plot(data_train)
plt.title("Training Data")
plt.xlabel("Time Index")
plt.ylabel("Price")
```

```python
# define A and b
A = np.zeros((3, 3))
b = np.zeros(3)

for ii in range(len(data_train)):
    if ii+3 >= len(data_train):
        break
    else:
        xt = data_train[ii:ii+3]
        yt = data_train[ii+3]
        A += np.outer(xt,xt)
        b += yt * xt
```

```python
A
```

```python
b
```

```python
w = np.linalg.inv(A) @ b
w
```

```python
def model(a_vec, xt_prev):
    return np.dot(a_vec, xt_prev)
```

```python
pred=np.zeros(len(data_test))
```

```python
data_train[-3:-1]
```

```python
for ii in range(len(data_test)):
    if ii==0:
        pred[ii]= model(a_vec=w, xt_prev=data_train[-4:-1])
    elif ii==1:
        pred[ii]= model(a_vec=w, 
                        xt_prev=np.array([data_train[-2], data_train[-1], data_test[0]]))
    elif ii==2:
        pred[ii]= model(a_vec=w, 
                        xt_prev=np.array([data_train[-1], data_test[0], data_test[1]]))
    else:
        pred[ii]= model(a_vec=w, 
                        xt_prev=data_test[ii-3:ii])
```

```python
plt.plot(pred, label="model")
plt.plot(data_test, label="test data")
plt.legend()
```

```python
np.sqrt(np.sum((pred - data_test)**2)/len(data_test))
```

```python
pred_train = np.ones(len(data_train))*np.infty

for ii in range(3, len(data_train)):
    pred_train[ii]= model(a_vec=w, 
                        xt_prev=data_train[ii-3:ii])
```

```python
plt.plot(pred_train, label="model")
plt.plot(data_train, label="test data")
plt.legend()
```

```python
np.sqrt(np.sum((pred_train[4:] - data_train[4:])**2)/(len(data_train)-3))
```

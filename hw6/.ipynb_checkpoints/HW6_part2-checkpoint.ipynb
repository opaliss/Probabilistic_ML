```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
import matplotlib.pyplot as plt
import matplotlib
font = {'family' : 'serif',
        'size'   : 15}

matplotlib.rc('font', **font)
matplotlib.rc('xtick', labelsize=15) 
matplotlib.rc('ytick', labelsize=15) 
```

```python
def f(x):
    return np.log(np.cosh(x))

def f_prime(x):
    return np.tanh(x)
```

```python
def Q(x, y):
    return f(y) + f_prime(y)*(x-y) + 0.5 *(x-y)**2
```

```python
x_vec = np.linspace(-10, 10, 10000)


fig, ax = plt.subplots(figsize=(8, 4))

ax.plot(x_vec, f(x_vec), label="f(x)")
ax.plot(x_vec, Q(x_vec, -2), label="Q(x, -2)")
ax.plot(x_vec, Q(x_vec, 3), label="Q(x, 3)")
ax.set_xlabel("x")
ax.set_title("6.3 (c) Auxiliary Function")
plt.legend()
plt.savefig("aux_function.png")
```

```python
n_vec = np.arange(0, 50)
x_vec_2 = np.zeros(len(n_vec))
x_vec_3 = np.zeros(len(n_vec))
```

```python
x_vec_2[0] = -2
x_vec_3[0] = 3
for ii in n_vec:
    if ii !=49:
        x_vec_2[ii+1] = x_vec_2[ii] - f_prime(x_vec_2[ii])
        x_vec_3[ii+1] = x_vec_3[ii] - f_prime(x_vec_3[ii])
```

```python
plt.plot(n_vec, x_vec_2, label="x0 = -2")
plt.plot(n_vec, x_vec_3, label="x0 = 3")
plt.legend()
plt.xlabel("n (iteration)")
plt.ylabel("$x_{n}$")
plt.title("Convergence to minimum")
```

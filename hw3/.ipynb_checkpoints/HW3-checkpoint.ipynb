# 3.5 (b)

```python
nsamples = int(1e6)
Bsamples = np.random.choice([0, 1], (10, nsamples))
```

```python
Bsamples.shape
```

```python
def indicator(pos, a):
    if a[pos] == 1: 
        return 1
    else:
        return 0
```

```python
def func(B):
    sol = 0
    for ii in range(1, 11):
        sol += 2**(ii-1)*B[ii-1]
    return sol
```

```python
def p_dist(f, alpha=0.1, z=128):
    return ((1 - alpha)/(1 + alpha)) * (alpha ** np.abs(z - f))
```

```python
def compute_prob(nsamples,pos, Bsamples):
    num = 0
    denom = 0
    for ii in range(nsamples):
        val = p_dist(f=func(B=Bsamples[:, ii]))
        denom += val
        if indicator(pos, Bsamples[:, ii]):
            num += val
    return num/denom
```

```python
print("Answer to 3.5(b) with a million samples")
print("P(B2|Z=128) = ", compute_prob(nsamples=nsamples, pos=1, Bsamples=Bsamples))
print("P(B5|Z=128) = ", compute_prob(nsamples=nsamples, pos=4, Bsamples=Bsamples))
print("P(B8|Z=128) = ", compute_prob(nsamples=nsamples, pos=7, Bsamples=Bsamples))
print("P(B10|Z=128) = ", compute_prob(nsamples=nsamples, pos=9, Bsamples=Bsamples))
```

# 3.5 (c)

```python
nsamples_vec = 10**np.linspace(1, 7, 6)
len(nsamples_vec)
```

```python
Bsamples = np.random.choice([0, 1], (10, 10**7))
```

```python
np.shape(Bsamples)
```

```python
res2 = np.zeros(len(nsamples_vec))
res5 = np.zeros(len(nsamples_vec))
res8 = np.zeros(len(nsamples_vec))
res10 = np.zeros(len(nsamples_vec))

kk=0
for jj in range(len(nsamples_vec)):
    print(kk)
    res2[kk] = compute_prob(nsamples=int(nsamples_vec[kk]), pos=1, Bsamples=Bsamples[:, :int(nsamples_vec[kk])])
    res5[kk] = compute_prob(nsamples=int(nsamples_vec[kk]), pos=4, Bsamples=Bsamples[:, :int(nsamples_vec[kk])])
    res8[kk] = compute_prob(nsamples=int(nsamples_vec[kk]), pos=7, Bsamples=Bsamples[:, :int(nsamples_vec[kk])])
    res10[kk] = compute_prob(nsamples=int(nsamples_vec[kk]), pos=9, Bsamples=Bsamples[:, :int(nsamples_vec[kk])])
    kk+=1 
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
jj=2
ax.plot(nsamples_vec[jj:], res2[jj:], label="P($B_{2} = 1| Z=128$)")
ax.plot(nsamples_vec[jj:], res5[jj:], label="P($B_{5} = 1| Z=128$)")
ax.plot(nsamples_vec[jj:], res8[jj:], label="P($B_{8} = 1| Z=128$)")
ax.plot(nsamples_vec[jj:], res10[jj:], label="P($B_{10} = 1| Z=128$)")
ax.set_xlabel("number of samples", fontsize=20)
ax.set_ylabel("conditional probability ", fontsize=20)
plt.legend()
ax.set_xscale("log")
```

```python
fig, ax = plt.subplots(figsize=(10, 5))


ax.plot(nsamples_vec[jj:-1], np.abs(res2[jj+1:] - res2[jj:-1]), label="P($B_{2} = 1| Z=128$)")
ax.plot(nsamples_vec[jj:-1], np.abs(res5[jj+1:] - res5[jj:-1]), label="P($B_{5} = 1| Z=128$)")
ax.plot(nsamples_vec[jj:-1], np.abs(res8[jj+1:]- res8[jj:-1]), label="P($B_{8} = 1| Z=128)$")
ax.plot(nsamples_vec[jj:-1], np.abs(res10[jj+1:]- res10[jj:-1]), label="P($B_{10} = 1| Z=128$)")
ax.set_xlabel("number of samples", fontsize=20)
ax.set_ylabel("cnvergence rate ", fontsize=20)

plt.legend()
ax.set_xscale("log")
```

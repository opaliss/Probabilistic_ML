```python
def log_likelihood(ratings, prob_r, prob_z):
    T = np.shape(ratings)[0] # number of students in class
    L = 0
    for tt in range(T):
        P_holder = 0
        for ii in range(0, 4):
            P_holder += prob_z[ii]*p_r_given_z(ratings, ii, tt)
        L += np.log(P_holder)
    return L/T

def p_r_given_z(ratings, ii, tt):
    holder=1
    set_movies_seen = np.append(np.where(ratings[tt, :] == "1")[0], np.where(ratings[tt, :] == "0")[0])
    for jj in set_movies_seen:
        if ratings[tt, jj] =="0":
            holder*=(1 - prob_r[jj, ii])
        elif ratings[tt, jj] == "1":
            holder*=prob_r[jj, ii]
    return holder

def e_step(ratings, prob_r, prob_z, ii, tt):
    num = prob_z[ii]*p_r_given_z(ratings=ratings, ii=ii, tt=tt)
    denom = 0
    for kk in range(0, 4):
        denom+= prob_z[kk]*p_r_given_z(ratings=ratings, ii=kk, tt=tt)
    return num/denom

def indicator(a, b):
    if a==b:
        return 1
    else:
        return 0
```

```python
def update_em(ratings, prob_r, prob_z):
    T = np.shape(ratings)[0]
    num_movies = np.shape(ratings)[1]
    prob_z_updated = np.zeros(np.shape(prob_z))
    prob_r_updated = np.zeros(np.shape(prob_r))
    rho_it = np.zeros((4, T))
    
    # prob z 
    for ii in range(0, 4):
        holder=0
        for tt in range(T):
            rho_it[ii, tt] = e_step(ratings=ratings, prob_r=prob_r, prob_z=prob_z, ii=ii, tt=tt)
            holder += rho_it[ii, tt]
        prob_z_updated[ii] = holder/T
    
    # prob r
    for ii in range(0, 4):
        for jj in range(num_movies):
            num = 0 
            for tt in range(T):
                set_of_seen_movies = np.append(np.where(ratings[tt, :] == "1")[0], np.where(ratings[tt, :] == "0")[0])
                if jj in set_of_seen_movies:
                    num+= rho_it[ii, tt] * indicator(ratings[tt, jj], "1")
                    
                else:
                    num+= rho_it[ii, tt] * prob_r[jj, ii]

                        
            prob_r_updated[jj, ii] = num/(prob_z_updated[ii]*T)
            
    return prob_z_updated, prob_r_updated, rho_it
```

```python
for ii in range(0, 257):
    if ii in [0, 1, 2, 4, 8, 16, 32, 64, 128, 256]:
        print("iteration " + str(ii) + ", log-likelihood = ", log_likelihood(ratings=ratings, prob_r=prob_r, prob_z=prob_z))
        print("\n")
    
    # update cpts 
    prob_z, prob_r, rho_it = update_em(ratings=ratings, prob_r=prob_r, prob_z=prob_z)
```

```python
indx_pid = np.where(ids=="A14974360")[0][0]-1
indx_pid
```

```python
not_seen = np.where(ratings[indx_pid, :]=="?")[0]

prob_not_seen = np.zeros(len(not_seen))
```

```python
def prob_not_seen_func(rho_it, tt, jj, prob_r):
    p = 0
    for ii in range(0, 4):
        p+=rho_it[ii, tt]*prob_r[jj, ii]
    return p
```

```python
for ii in range(len(not_seen)):
    prob_not_seen[ii] = prob_not_seen_func(rho_it=rho_it, tt=indx_pid, jj=not_seen[ii], prob_r=prob_r)
```

```python
print(movies[np.flip(np.argsort(prob_not_seen))])
```
